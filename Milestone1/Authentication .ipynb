```python
import smtplib, re
import random
from email.mime.text import MIMEText

sender = "dhanushadesu@gmail.com"
app_password = "YOUR_APP_PASSWORD"

# Email Validation Function

def is_valid_email(email):
    pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
    return re.fullmatch(pattern, email) is not None

# Keep Asking Until Valid Email

while True:
    receiver = input("Enter your Email: ").strip()

    if is_valid_email(receiver):
        break
    else:
        print("❌ Invalid Email Address! Please enter a valid email.\n")

otp = random.randint(100000,999999)

msg = MIMEText(f"Your OTP is {otp}")
msg["Subject"] = "MoodMentor OTP"
msg["From"] = sender
msg["To"] = receiver

server = smtplib.SMTP("smtp.gmail.com",587)
server.starttls()
server.login(sender, app_password)
server.sendmail(sender, receiver, msg.as_string())
server.quit()

print("OTP Sent!")
print("OTP =", otp)

entered_otp = input("Enter the OTP received in your email: ")

if entered_otp == str(otp):
    print("✅ Email Verified Successfully!")
else:
    print("❌ Invalid OTP")
```

```python
import jwt
from datetime import datetime, timedelta, UTC

SECRET_KEY = "YOUR_SECRET_KEY"
# For HS256, the recommended secret key length is at least 32 bytes.

payload = {
    "email": receiver,
    "exp": datetime.now(UTC) + timedelta(hours=1)
}

token = jwt.encode(payload, SECRET_KEY, algorithm="HS256")

print("JWT Token:")
print(token)
```

```python
try:
    decoded = jwt.decode(token, SECRET_KEY, algorithms=["HS256"])
    print("Token is Valid")
    print(decoded)
except jwt.ExpiredSignatureError:
    print("Token Expired")
except jwt.InvalidTokenError:
    print("Invalid Token")
```
